            <li>Variable names may only start with a letter or underscore (special use cases)</li>
            <li>There are two common naming conventions for avoid spaces:</li>
            <ul>
                <li>snake_case - spaces are replaced with underscores</li>
                <li>cammelCase - words are joined with capital letters at the start of every word (not including the first word - that is called PascalCase)</li>
            </ul>
            <li>The standard for Python is noramlly snake case - but in general as long as you are consistent in your program with which ever you use, thats fine!</li>

    <div class="sub-section">
        <h3>4.4 - Data Types</h3>
        <p>We have seen variables in use, we have seen the operators that can be used on them. We have even mentioned data types above once or twice already, so what are they?</p>
        <p>Data types specify what kind of data is being stored or used in a program, for now there are four we will concern ourselves with (but there are more!).</p>
        <ol>
            <li>int - short for integer, a whole positive or negative number</li>
            <li>float - sometimes called a real, a positive or negative number with a decimal place (note this includes things like 1.0)</li>
            <li>string - plain text, one of the most common data types used, denoted with 'single' or "double" quotes</li>
            <li>bool - short for Boolean, takes either the value <span class="code">True</span> or <span class="code">False</span> - used for decision making in programs</li>
        </ol>
        <p>Note - Python is refered to as a dynamically typed langauge and a strongly typed langauge.</p>
        <ul>
            <li>Dynamic - you do not tell Python what kind of data a variable will hold, Python figures this out for you! This also means that you can change the data type a variable holds.</li>
            <li>Strongly typed - Python will limit combining data types in operations - more below.</li>
        <p>Python has a built in function for finding the data type of something, it is called <span class="code">type()</span></p>
    </div>
</div>

```python
print(type(1)) # --> int
print(type(1.5)) # --> float
print(type(1.0)) # --> float
print(type("1")) # --> str
print(type(True)) # --> bool
print(type("True")) # --> str
print(type(5 > 2)) # --> bool
```
